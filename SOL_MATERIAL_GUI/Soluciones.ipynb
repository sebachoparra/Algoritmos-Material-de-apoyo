## Ejercicio 1: Clasificación de Números
- Crea una interfaz que permita ingresar un número y clasificarlo como positivo, negativo o cero.
```python
import gradio as gr
```
```python
def clasificar_numero(num):
    if num > 0:
        return "Número positivo"
    elif num < 0:
        return "Número negativo"
    else:
        return "Es cero"
```
```python
iface = gr.Interface(fn=clasificar_numero, inputs=gr.Number(label="Ingrese un número"), outputs="text")
iface.launch(share=True)
```

## Ejercicio 2: Calculadora de Descuentos
- Ingresar un precio y un porcentaje de descuento, mostrar el precio final.
```python
import gradio as gr
```
```python
def calcular_descuento(precio, descuento):
    precio_final = precio - (precio * descuento / 100)
    return f"Precio final: ${precio_final:.2f}"
```
```python
iface = gr.Interface(
    fn=calcular_descuento,
    inputs=[gr.Number(label="Precio"), gr.Slider(0, 100, step=1, label="% Descuento")],
    outputs="text")
iface.launch(share=True)
```

## Ejercicio 3: Suma de Números Pares
- Ingresar una lista de números y devolver la suma de los pares.
```python
import gradio as gr
```
```python
def sumar_pares(lista):
    numeros = [int(x) for x in lista.split(",")]
    suma = sum(x for x in numeros if x % 2 == 0)
    return f"Suma de números pares: {suma}"
```
```python
iface = gr.Interface(fn=sumar_pares, inputs=gr.Textbox(label="Lista de números separados por coma"), outputs="text")
iface.launch(share=True)
```

## Ejercicio 8: Análisis de Listas
- Ingresar una lista y mostrar promedio, máximo y mínimo.
```python
import gradio as gr
```
```python
def analizar_lista(lista):
    numeros = [float(x) for x in lista.split(",")]
    return {
        "Promedio": sum(numeros) / len(numeros),
        "Máximo": max(numeros),
        "Mínimo": min(numeros)
    }
```
```python
iface = gr.Interface(fn=analizar_lista, inputs=gr.Textbox(label="Números separados por coma"), outputs="label")
iface.launch(share=True)
```

## Ejercicio 9: Operaciones con Matrices
- Suma de dos matrices ingresadas.
```python
import gradio as gr
import numpy as np
```
```python
def sumar_matrices(matriz1, matriz2):
    m1 = np.array(matriz1)
    m2 = np.array(matriz2)
    return (m1 + m2).tolist()
```
```python
iface = gr.Interface(
    fn=sumar_matrices,
    inputs=[
        gr.Dataframe(row_count=3, col_count=3, label="Matriz 1"),
        gr.Dataframe(row_count=3, col_count=3, label="Matriz 2")
    ],
    outputs=gr.Dataframe()
)
iface.launch(share=True)
```

## Ejercicio 10: Determinante e Inversa de una Matriz
- :)
```python
import gradio as gr
import numpy as np
```
```python
def det_inversa(matriz):
    m = np.array(matriz)
    determinante = np.linalg.det(m)
    try:
        inversa = np.linalg.inv(m).tolist()
    except:
        inversa = "La matriz no tiene inversa"
    return {"Determinante": determinante, "Inversa": inversa}
```
```python
iface = gr.Interface(
    fn=det_inversa,
    inputs=gr.Dataframe(row_count=3, col_count=3, label="Matriz"),
    outputs="json"
)
iface.launch(share=True)
```
